# What is the probability that you draw a wild card or a +2 card?
The sample space is all 108 cards in the Uno deck. One trial is drawing one cart and get either a wild card or a +2 of any color. The outcome we are measuring is whether the card is a Wild or +2 card. The probability of this outcome is 8/108 + 2/108 + 2/108 + 2/108 + 2/108 = 16/108 or 4/27

```python
import random
import matplotlib.pyplot as plt

# Define the Uno deck
deck = (
    ['Wild'] * 8 +                # 8 Wild cards
    ['+2 Red'] * 2 +              # 2 +2 cards of each color
    ['+2 Blue'] * 2 +
    ['+2 Green'] * 2 +
    ['+2 Yellow'] * 2 +
    ['Number/Other'] * (108 - 16) # All other cards
)

# Cards we are tracking
special_cards = ['Wild', '+2 Red', '+2 Blue', '+2 Green', '+2 Yellow']

# Initialize results dictionary
results = {card: 0 for card in special_cards}
results['Other'] = 0

# Run 100 simulations
trials = 100
for _ in range(trials):
    card = random.choice(deck)
    if card in special_cards:
        results[card] += 1
    else:
        results['Other'] += 1

# Print results with counts and percentages
print("Simulation Results (100 Trials):")
for card, count in results.items():
    percentage = (count / trials) * 100
    print(f"{card}: {count} times, {percentage:.1f}%")

# Display bar chart
plt.bar(results.keys(), results.values(), color=['orange','red','blue','green','yellow','grey'])
plt.ylabel('Frequency')
plt.title('Simulation of Drawing Wild or +2 Cards (100 Trials)')
plt.show()
```

In the simulation, 22% of the draws resulted in a Wild or +2 card, with Wild cards being the most common since there are eight of them in the deck. The theoretical probability for drawing a Wild or +2 card is 14%. This indicates that repeated trials can lead to slightly more variability in the outcomes than expected.
