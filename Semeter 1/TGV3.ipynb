# What is the probability that you draw a wild card and a red +2 card in your second draw?
The sample space is all 108 cards in the Uno deck. One trial is drawing two cards without replacement. The outcome we are measuring is whether the first card is a Wild and the second card is a red +2. The probability of this outcome is 8/108 * 2/108 = 16/11556

```python
import random
import matplotlib.pyplot as plt

# Build Uno deck
# Represent cards as strings
deck = []

# Add 8 wilds
deck += ["Wild"] * 4
deck += ["Wild Draw Four"] * 4

# Add +2 cards: 2 of each color
colors = ["Red", "Blue", "Green", "Yellow"]
for color in colors:
    deck += [f"{color} +2"] * 2

# Fill the rest of the deck with "Other" cards to reach 108
while len(deck) < 108:
    deck.append("Other")

# Simulation
trials = 100
success = 0
fail = 0

for _ in range(trials):
    drawn = random.sample(deck, 2)  # draw 2 cards without replacement
    if drawn[0] in ["Wild", "Wild Draw Four"] and drawn[1] == "Red +2":
        success += 1
    else:
        fail += 1

# Estimated probability
probability = success / trials
print(f"Estimated probability after {trials} trials: {probability:.4f}")

# Bar chart
plt.bar(["Success (Wild then Red +2)", "Failure"], [success, fail], color=["red", "blue"], edgecolor="black")
plt.title("Simulation: Drawing a Wild then a Red +2 (100 Trials)")
plt.ylabel("Count")
plt.show()
```

Both simulations and theoretical analysis show that drawing a wild card followed by a red +2 is very rare. In 1,000 simulation trials, the probability was 0.001%, while the theoretical probability is about 0.0013845 (16/11,664). Both indicate that this event is highly unlikely.
