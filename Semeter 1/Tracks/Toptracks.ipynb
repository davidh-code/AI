```python
import pandas as pd

# Load your dataset
df = pd.read_csv("global.csv")

# Split "Wiki Genre" into individual genres
wiki_genres = df["Wiki Genre"].dropna().str.split(", ")

# Flatten and take only the first word of each genre
all_genres = [genre.split()[0] for sublist in wiki_genres for genre in sublist]

# Normalize to lowercase
all_genres = [g.lower() for g in all_genres]

# Create frequency table
wiki_genre_freq = pd.Series(all_genres).value_counts().reset_index()
wiki_genre_freq.columns = ["Genre", "Frequency"]

# Add relative frequency (percentage)
total = wiki_genre_freq["Frequency"].sum()
wiki_genre_freq["Relative Frequency"] = (wiki_genre_freq["Frequency"] / total * 100).round(2)

# Print table
print(wiki_genre_freq)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("global.csv")

# Split and clean genres
wiki_genres = df["Wiki Genre"].dropna().str.split(", ")
all_genres = [genre.split()[0].lower() for sublist in wiki_genres for genre in sublist]

# Frequency table
wiki_genre_freq = pd.Series(all_genres).value_counts().reset_index()
wiki_genre_freq.columns = ["Genre", "Frequency"]

# Limit to top 10, group rest into "Other"
top10 = wiki_genre_freq.head(10)
other_sum = wiki_genre_freq["Frequency"].iloc[10:].sum()
top10.loc[len(top10)] = ["Other", other_sum]

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(
    top10["Frequency"],
    labels=top10["Genre"],
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.8,   # move % closer to center
    labeldistance=1.1  # push labels outward
)
plt.title("Top 10 Genres (first word only)", fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Load dataset
df = pd.read_csv("global.csv")

# Split and clean genres
wiki_genres = df["Wiki Genre"].dropna().str.split(", ")
all_genres = [genre.split()[0].lower() for sublist in wiki_genres for genre in sublist]

# Frequency table
wiki_genre_freq = pd.Series(all_genres).value_counts().reset_index()
wiki_genre_freq.columns = ["Genre", "Frequency"]

# Limit to top 10, group rest into "Other"
top10 = wiki_genre_freq.head(10)
other_sum = wiki_genre_freq["Frequency"].iloc[10:].sum()
top10.loc[len(top10)] = ["Other", other_sum]

# Generate unique colors using a colormap
colors = cm.tab20.colors[:len(top10)]  # pick N distinct colors

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(
    top10["Frequency"],
    labels=top10["Genre"],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,        # apply unique colors
    pctdistance=0.8,      # move % closer to center
    labeldistance=1.1     # push labels outward
)
plt.title("Top 10 Genres (first word only)", fontsize=14)
plt.show()
```
