```python
import pandas as pd

# Load the Pok√©mon dataset
df = pd.read_csv("Pokemon.csv")

# Keep only base form Pok√©mon (those without "Evolves From")
base_pokemon = df[df["Evolves From"].isna()]

# Extract Attack values
attack_values = base_pokemon["Attack"]

# Calculate measures of central tendency
mean_attack = attack_values.mean()
median_attack = attack_values.median()
mode_attack = attack_values.mode().tolist()  # could be more than one mode

print("Measures of Central Tendency for Attack (Base Form Pok√©mon Only):")
print(f"Mean:   {mean_attack:.2f}")
print(f"Median: {median_attack}")
print(f"Mode:   {mode_attack}")
```

```python
import pandas as pd

# Load the Pok√©mon dataset
df = pd.read_csv("Pokemon.csv")

# Keep only base form Pok√©mon (those without "Evolves From")
base_pokemon = df[df["Evolves From"].isna()]

# Extract Attack values
attack_values = base_pokemon["Attack"]

# Calculate measures of spread
attack_range = attack_values.max() - attack_values.min()
attack_variance = attack_values.var()
attack_std = attack_values.std()
attack_iqr = attack_values.quantile(0.75) - attack_values.quantile(0.25)

print("Measures of Spread for Attack (Base Form Pok√©mon Only):")
print(f"Range: {attack_range}")
print(f"Variance: {attack_variance:.2f}")
print(f"Standard Deviation: {attack_std:.2f}")
print(f"IQR (Interquartile Range): {attack_iqr}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Pok√©mon dataset
df = pd.read_csv("Pokemon.csv")

# Keep only base form Pok√©mon (those without "Evolves From")
base_pokemon = df[df["Evolves From"].isna()]

# Extract Attack values
attack_values = base_pokemon["Attack"]

# Create box plot
plt.figure(figsize=(6, 8))
plt.boxplot(attack_values, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot of Attack (Base Form Pok√©mon Only)")
plt.ylabel("Attack Value")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Pokemon.csv")

# Keep only base form Pok√©mon (those without "Evolves From")
base_pokemon = df[df["Evolves From"].isna()]

# Sort base Pok√©mon by Attack (descending = strongest first)
sorted_attack = base_pokemon.sort_values(by="Attack", ascending=False).reset_index(drop=True)

# Find Charmander's position
charmander_position = sorted_attack[sorted_attack["Name"] == "Charmander"].index[0] + 1  # +1 to make it human-readable rank

# Total number of base form Pok√©mon
total_base = len(sorted_attack)

print(f"Charmander's Attack rank: {charmander_position} out of {total_base} base form Pok√©mon")
```
