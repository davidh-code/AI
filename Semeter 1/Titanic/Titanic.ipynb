```python
import pandas as pd
df = pd.read_excel("table.xlsx")

# Filter only the United Kingdom row
uk_data = df[df["Country of Origin"] == "United Kingdom"]

# Extract the passenger counts by class
first_class = uk_data["First Class Passenger Count by Country"].values[0]
second_class = uk_data["Second Class Passenger Count by Country"].values[0]
third_class = uk_data["Third Class Passenger Count by Country"].values[0]

print("British Passengers by Class:")
print("  First Class:", first_class)
print("  Second Class:", second_class)
print("  Third Class:", third_class)

total_passengers = df["Total Passenger Count by Country"].sum()
print("Total passengers on the Titanic:", total_passengers)

import matplotlib.pyplot as plt

# Set the index to Country for easier plotting
df.set_index("Country of Origin")[[
    "First Class Passenger Count by Country",
    "Second Class Passenger Count by Country",
    "Third Class Passenger Count by Country"
]].plot(kind="bar", stacked=True, figsize=(14,7))

plt.title("Titanic Passengers by Nationality and Class")
plt.xlabel("Country of Origin")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
