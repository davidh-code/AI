```python
import pandas as pd

# Load your dataset
df = pd.read_csv("ramen.csv")

# Calculate Pearson correlation between Yelp Rating and Yelp Reviews
r = df['Yelp Rating'].corr(df['Yelp Reviews'])

# Coefficient of Determination (R²)
r_squared = r**2

print("Pearson correlation coefficient (r):", r)
print("Coefficient of Determination (R²):", r_squared)
```

```python
import pandas as pd
import numpy as np

# Load your CSV file
df = pd.read_csv("ramen.csv")

# Make sure we only use rows that have both Yelp Reviews and Yelp Rating
df = df.dropna(subset=["Yelp Reviews", "Yelp Rating"])

# Extract the data
x = df["Yelp Reviews"]
y = df["Yelp Rating"]

# Calculate correlation coefficient (r)
r = x.corr(y)

# Calculate coefficient of determination (R²)
r2 = r ** 2

print(f"Correlation (r): {r:.4f}")
print(f"Coefficient of determination (R²): {r2:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv("ramen.csv")

# Keep only rows with both Yelp Reviews and Yelp Rating
df = df.dropna(subset=["Yelp Reviews", "Yelp Rating"])

# Extract values
x = df["Yelp Reviews"]
y = df["Yelp Rating"]

# Calculate correlation coefficient (r) and R²
r = x.corr(y)
r2 = r ** 2

# Fit regression line (y = m*x + b)
m, b = np.polyfit(x, y, 1)
line = m * x + b

# Create scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.7, color="blue")  # just dots

# Plot regression line
plt.plot(x, line, color="red", linewidth=2, label="Best Fit Line")

# Labels and title
plt.xlabel("Yelp Reviews")
plt.ylabel("Yelp Rating")
plt.title("Yelp Rating vs Yelp Reviews (All Restaurants)")

# Add r and R² text box inside the plot
plt.text(
    0.05, 0.95,
    f"Correlation (r): {r:.4f}\nR²: {r2:.4f}",
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment="top",
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white")
)

plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
