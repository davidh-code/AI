# Given that the first card drawn is a special card (either wild or action), what is the probability that it is a wild card?

The sample space includes all special cards in UNO, which consists of both wild and action cards, totaling 20 special cards. In this simulation, one trial involves drawing a random special card from the deck. The outcome being measured is whether the drawn card is a wild card. There are 8 wild cards among the 20 special cards, so the probability of drawing a wild card is 2/5.

```python
import random
import matplotlib.pyplot as plt

# Uno deck setup
total_cards = 108
wild_cards = 8
action_cards = 24   # Skip, Reverse, Draw Two (24 total)
special_cards = wild_cards + action_cards

# Simulation parameters
trials = 100
special_draws = 0
wild_given_special = 0
action_given_special = 0

for _ in range(trials):
    drawn_card = random.randint(1, total_cards)

    if drawn_card <= special_cards:  # special card drawn
        special_draws += 1
        if drawn_card <= wild_cards:  # wild card
            wild_given_special += 1
        else:  # action card
            action_given_special += 1

# Conditional probability estimate
if special_draws > 0:
    estimated_probability = wild_given_special / special_draws
    print(f"Out of {special_draws} specials, {wild_given_special} were wilds and {action_given_special} were actions.")
    print(f"Estimated P(Wild | Special) after {trials} runs: {estimated_probability:.4f}")
else:
    print("No special cards were drawn in the 100 runs.")

# --- Bar chart ---
labels = ["Wild", "Action"]
counts = [wild_given_special, action_given_special]

plt.bar(labels, counts)
plt.title("Special Cards Distribution (100 Runs)")
plt.ylabel("Count")
plt.show()
```

In one hundred simulations, a special card appeared in the first move 28 times. Among these instances, 19 were action cards and 9 were wild cards. This means that the percentage of drawing a wild card as a special card in the first move is 32.14%. Theoretically, this percentage would be 40%, as there are 4 wild cards out of a total of 10 special cards.
